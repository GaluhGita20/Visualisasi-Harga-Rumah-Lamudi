```python
import pandas as pd

import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
```

```python
df_house=pd.read_csv('C:\House Prediction\House-Price-Prediction\dataset\house_predict.csv')
df_house.info()
```

```python
df_house=df_house[['harga','parkir motor','kamar tidur','kamar mandi','jmlh lantai','luas bangunan','luas tanah', 'kabupaten']]
dict_name={'parkir motor': 'parkir_motor',
           'kamar tidur':'kamar_tidur',
           'kamar mandi':'kamar_mandi',
           'jmlh lantai':'jumlah_lantai',
           'luas bangunan':'luas_bangunan',
           'luas tanah':'luas_tanah'
}
df_house.rename(columns=dict_name, inplace=True)
df_house.head()
```

```python
df_house['kabupaten'].unique()
```

```python
house_karangasem=df_house[df_house['kabupaten']=='karangasem'] 
house_gianyar=df_house[df_house['kabupaten']=='gianyar']    
house_badung=df_house[df_house['kabupaten']=='badung']
house_tabanan=df_house[df_house['kabupaten']=='tabanan']
house_buleleng=df_house[df_house['kabupaten']=='buleleng']
house_jembrana=df_house[df_house['kabupaten']=='jembrana']
house_bangli=df_house[df_house['kabupaten']=='bangli']
house_denpasar=df_house[df_house['kabupaten']=='denpasar']
```

```python
lm_karangasem=smf.ols(formula='harga ~ kamar_tidur + kamar_mandi + jumlah_lantai + luas_bangunan + luas_tanah', data=house_karangasem).fit()
lm_gianyar=smf.ols(formula='harga ~ kamar_tidur + kamar_mandi + jumlah_lantai + luas_bangunan + luas_tanah', data=house_gianyar).fit()
lm_badung=smf.ols(formula='harga ~  kamar_tidur + kamar_mandi + jumlah_lantai + luas_bangunan + luas_tanah', data=house_badung).fit()
lm_tabanan=smf.ols(formula='harga ~  kamar_tidur + kamar_mandi + jumlah_lantai + luas_bangunan + luas_tanah', data=house_tabanan).fit()
lm_buleleng=smf.ols(formula='harga ~  kamar_tidur + kamar_mandi + jumlah_lantai + luas_bangunan + luas_tanah', data=house_buleleng).fit()
lm_jembrana=smf.ols(formula='harga ~  kamar_tidur + kamar_mandi + jumlah_lantai + luas_bangunan + luas_tanah', data=house_jembrana).fit()
lm_bangli=smf.ols(formula='harga ~  kamar_tidur + kamar_mandi + jumlah_lantai + luas_bangunan + luas_tanah', data=house_bangli).fit()
lm_denpasar=smf.ols(formula='harga ~  kamar_tidur + kamar_mandi + jumlah_lantai + luas_bangunan + luas_tanah', data=house_denpasar).fit()
```

```python
#uji multicolinierity
from sklearn.preprocessing import MinMaxScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor
def preprocessing(X):    
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)
    variables = X_scaled
    vif = pd.DataFrame()
    vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
    vif["Features"] = X.columns
    print(vif)
```
